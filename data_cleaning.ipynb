REMOVE UNEQUAL LISTS

```python
import csv

def isEqual(odd_list, even_list):
    return (len(odd_list) == len(even_list))


def writeToCSV(list):
    # Write the output data to a new CSV file
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(list)

def filter_csv(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        total_rows = 0
        removed_rows = 0
        filtered_rows = 0

        total_lists = 0
        removed_lists = 0
        filtered_lists = 0

        shopper_num_1 = 0
        shopper_num_2 = 0
        invalid_shoppers = []

        odd_list = []
        even_list = []

        main_list = []
        list_flag = 0

        for row in reader:

            # for ODD list
            if float(row[1]) == 0:
                if int(row[2]) % 2 == 1:
                    shopper_num_1 = row[2]
                    # if new shopper
                    if odd_list and even_list:

                        if isEqual(odd_list, even_list):
                            main_list.extend(odd_list)
                            main_list.extend(even_list)
                            total_lists += 2
                        else:
                            invalid_shoppers.append(shopper_num_1)
                            invalid_shoppers.append(shopper_num_2)
                            filtered_lists += 2

                        odd_list.clear()
                        even_list.clear()

                    odd_list.append(row)
                    list_flag = 1

            else:
                if list_flag == 1:
                    odd_list.append(row)

            # for EVEN list
            if float(row[1]) == 0:
                if int(row[2]) % 2 == 0:
                    shopper_num_2 = row[2]
                    even_list.append(row)
                    list_flag = 0
            else:
                if list_flag == 0:
                    even_list.append(row)

        # Check if the last row belongs to an odd or even list and process accordingly
        if list_flag == 1:
            odd_list.append(row)
        elif list_flag == 0:
            even_list.append(row)

        # Process the last odd and even lists
        if odd_list and even_list:
            if isEqual(odd_list, even_list):
                main_list.extend(odd_list)
                main_list.extend(even_list)
                total_lists += 2
            else:
                invalid_shoppers.append(shopper_num_1)
                invalid_shoppers.append(shopper_num_2)
                filtered_lists += 2

        # Clear the lists after processing the last row
        odd_list.clear()
        even_list.clear()


        removed_lists = abs(total_lists - filtered_lists)

        writeToCSV(main_list)

        print(f"Total Lists: {total_lists}")
        print(f"Total Lists Removed: {removed_lists}")
        print(f"Total Filtered Lists: {filtered_lists}")

        invalid_shoppers_str = ', '.join(str(shopper) for shopper in invalid_shoppers)

        # Print the list
        print(f"Total Invalid Shoppers: {len(invalid_shoppers)}")
        print(f"Invalid Shoppers (no.): {invalid_shoppers_str}")


if __name__ == "__main__":
    input_file = "1-raw.csv"
    output_file = "2-unequal-removed.csv"

    filter_csv(input_file, output_file)
```

```python
import csv

def isStuckHit(odd_list, even_list):
    if (odd_list and any(item[2] == "stuck hit!" for item in odd_list)) or (even_list and any(item[2] == "stuck hit!" for item in even_list)):
        return True

    return False

def writeToCSV(list, output_file):
    # Write the output data to a new CSV file
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(list)

def filter_csv(input_file, output_file):
    with open(input_file, 'r') as infile:
        reader = csv.reader(infile)

        total_rows = 0
        removed_rows = 0
        filtered_rows = 0

        total_lists = 0
        removed_lists = 0
        filtered_lists = 0

        shopper_num_1 = 0
        shopper_num_2 = 0
        invalid_shoppers = []

        odd_list = []
        even_list = []

        main_list = []
        list_flag = 0

        for row in reader:

            # for ODD list
            if float(row[1]) == 0:
                if int(row[2]) % 2 == 1:
                    shopper_num_1 = row[2]
                    # if new shopper
                    if odd_list and even_list:

                        if isStuckHit(odd_list, even_list):
                            invalid_shoppers.append(shopper_num_1)
                            invalid_shoppers.append(shopper_num_2)
                            removed_lists += 2
                        else:
                            main_list.extend(odd_list)
                            main_list.extend(even_list)
                            filtered_lists += 2

                        odd_list.clear()
                        even_list.clear()

                    odd_list.append(row)
                    list_flag = 1

            else:
                if list_flag == 1:
                    odd_list.append(row)

            # for EVEN list
            if float(row[1]) == 0:
                if int(row[2]) % 2 == 0:
                    shopper_num_2 = row[2]
                    even_list.append(row)
                    list_flag = 0
            else:
                if list_flag == 0:
                    even_list.append(row)

        # Process the last odd and even lists
        if odd_list and even_list:
            if isStuckHit(odd_list, even_list):
                invalid_shoppers.append(shopper_num_1)
                invalid_shoppers.append(shopper_num_2)
                removed_lists += 2
            else:
                main_list.extend(odd_list)
                main_list.extend(even_list)
                filtered_lists += 2

        # Clear the lists after processing the last row
        odd_list.clear()
        even_list.clear()

        total_lists = abs(removed_lists + filtered_lists)

        writeToCSV(main_list, output_file)

        print(f"Total Lists: {total_lists}")
        print(f"Total Lists Removed: {total_lists}")
        print(f"Total Filtered Lists: {filtered_lists}")

        invalid_shoppers_str = ', '.join(str(shopper) for shopper in invalid_shoppers)

        # Print the list
        print(f"Total Invalid Shoppers: {len(invalid_shoppers)}")
        print(f"Invalid Shoppers (no.): {invalid_shoppers_str}")


if __name__ == "__main__":
    input_file = "2-unequal-removed.csv"
    output_file = "3-stuckhit-removed.csv"

    filter_csv(input_file, output_file)
```

CONVERT VALIDATION_DATA TO INPUT

```python
import csv

input_file = "3-stuckhit-removed.csv"
output_file = "4-simplified.csv"

with open(input_file, newline='') as csvfile:
    reader = csv.reader(csvfile)
    
    current_list = []
    output_data = []
    
    for row in reader:
        if float(row[1]) == 0:
            # If the timestamp is 0, process the current list and reset
            if current_list:
                # Count the number of items in the list
                num_items = len(current_list)+1
                # Get the last timestamp of the list
                last_timestamp = current_list[-1][1]
                # Add the count and last timestamp to the output data
                output_data.append([num_items, last_timestamp])
                # Reset the current list
                current_list = []
        else:
            # If the timestamp is not 0, add the row to the current list
            current_list.append(row)

    # Process the last list if there is one
    if current_list:
        num_items = len(current_list)+1
        last_timestamp = current_list[-1][1]
        output_data.append([num_items, last_timestamp])

# Write the output data to a new CSV file
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(output_data)
```

```python
import pandas as pd
from tkinter import filedialog
import matplotlib.pyplot as plt

# file_path = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv")])
file_path = 'random_data.csv'
df = pd.read_csv(file_path, header=None)  # Assuming the CSV file has no header

df['TimePerItem_Algo1'] = df.iloc[:, 1] / df.iloc[:, 0]
df['TimePerItem_Algo2'] = df.iloc[:, 2] / df.iloc[:, 0]

avg_time_per_item_algo1 = df['TimePerItem_Algo1'].mean()
avg_time_per_item_algo2 = df['TimePerItem_Algo2'].mean()

percentage_difference = ((avg_time_per_item_algo2 - avg_time_per_item_algo1) / avg_time_per_item_algo1) * 100

print(f"Average Time per Item for Algo1: {avg_time_per_item_algo1:.2f}")
print(f"Average Time per Item for Algo2: {avg_time_per_item_algo2:.2f}")
print(f"ML-DProSA vs No Algorithm: {percentage_difference:.2f}%")

plt.figure(figsize=(10, 6))

plt.fill_between(df.index, df['TimePerItem_Algo1'], label='No algorithm', alpha=0.7, color='orange')
plt.fill_between(df.index, df['TimePerItem_Algo2'], label='ML-DProSA', alpha=0.5, color='green')

plt.title('ML-DProSA Performance')
plt.xlabel('Shopper')
plt.ylabel('Time per Item')
plt.legend()
plt.grid(True)

plt.show()
```
